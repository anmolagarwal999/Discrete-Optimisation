```python
drop_cols=['HT','ET','P','Comments','∑FT',"Group","Round"]
for i in drop_cols:
    df=df.drop([i], axis = 1) 
```

```python
def truncate_team_name(str1):
    str2=""
    for i in str1:
        if i=='›':
            break
        str2=str2+i
    str2=str2.replace(" ","_")
    print(str2)
    return str2
```

```python
#for i in len(df.index):
   # df.ix[i,'Team 1'] = truncate_team_name(df.ix[i,'Team 1'])
```

```python
df.index
```

```python
df['Team 1'] = df['Team 1'].map(lambda name: truncate_team_name(name))
df['Team 2'] = df['Team 2'].map(lambda name: truncate_team_name(name))
```

```python
df
```

```python
from collections import defaultdict

naive_score = defaultdict(dict)

advanced_score=defaultdict(dict)
```

```python
t1=[]
t2=[]
naive_score=[]
advanced_score=[]
```

```python
def add_edges(team1,team2,score_str):
    score_str=score_str.split()[0]
    #print(score_str)
    g1=int(score_str.split('-')[0])
    g2=int(score_str.split('-')[1])
    #print(g1)
    #print(g2)
    if team1=="FC_Barcelona_" and    team2=="Bayern_München_":
        print("SCORE IS ",score_str)
        
    if g1==g2:
        return
    if g1>g2:
        t1.append(team1)
        t2.append(team2)
        naive_score.append(1)
        advanced_score.append((g1-g2))
    if g2>g1:
        t1.append(team2)
        t2.append(team1)
        naive_score.append(1)
        advanced_score.append((g2-g1))
        
        
    
    
```

```python
for index, row in df.iterrows():
    add_edges(row["Team 1"],row["Team 2"],row["FT"])
   # print(row["Team 1"], row["Team 2"])
```

```python
#add_edges("a","b","3-1 ()")
```

```python
data_obj={"winner":t1,
          "loser":t2,
          "wt":naive_score,
          "adv_wt":advanced_score
    
}
```

```python
df2 = pd.DataFrame(data_obj) 
```
